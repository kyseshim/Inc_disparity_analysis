## Income Disparity Changes in South Korean Society:
A Comparative Analysis Before and After the COVID-19 Pandemic

### 1. Purpose

This project focuses on a comparative analysis before and after the COVID-19 pandemic of individuals' income (salary) and gender, ages, education level, occupation, and region in South Korea.

Not only in the world but also South Korean society, as the economy becomes difficult due to the COVID-19 pandemic, the income disparity widens, which can deepen the income inequality problem in South Korean society.

As the first part of creating a data set, the income gap will analyze by comparing individual income (salary) and demographic characteristic variables (gender, ages, education level, occupation, and region) by pre-processing and refining variables through Python programming with the Korean Welfare Panel Study data from 2016 to 2021 (6 years).

### 2. Methodolgy

### Data: Korean Welfare Panel Study data (2016-2021).
The data set will create by merging the Korean Welfare Panel Study data for six years (2016-2021) and selecting necessary variables

- Key variables: Income(salary), gender, age, education level, occupation, region.
- Analyzing differences through descriptive statistics, etc.

### 3. Data Preprocessing and Variables Refining

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Load Data
raw_welfare1 = pd.read_excel('data/Koweps_Merge files_2016_2018.xlsx')
raw_welfare2 = pd.read_excel('data/Koweps_Merge files_2019_2021.xlsx')
raw_welfare3 = pd.read_excel('data/Koweps_Merge files_2016_2021.xlsx')

# Make Copy
welfare1 = raw_welfare1.copy()
welfare2 = raw_welfare2.copy()
welfare3 = raw_welfare3.copy()
print(welfare1)
print(welfare2)
print(welfare3)
```

```python
welfare3 = pd.DataFrame(raw_welfare3)
welfare3
```

```python
# Remove missing values
# 2016-2018
welfare1 = pd.DataFrame(welfare1)
pd.isna(welfare1) # check missing values
pd.isna(welfare1).sum() # Frequency of missing value
welfare1 = welfare1.dropna(subset = ['ID', 'Year', 'Reg7', 'Gender', 'BirthYear', 'EduLevel', 'MarriType', 'Occupation', 'Income'])
welfare1
```

```python
# Remove missing values
# 2019-2021
welfare2 = pd.DataFrame(welfare2)
pd.isna(welfare2) # check missing values
pd.isna(welfare2).sum() # Frequency of missing value
welfare2 = welfare2.dropna(subset = ['ID', 'Year', 'Reg7', 'Gender', 'BirthYear', 'EduLevel', 'MarriType', 'Occupation', 'Income'])
welfare2
```

```python
# Remove missing values
# 2016-2021
welfare3 = pd.DataFrame(welfare3)
pd.isna(welfare3) # check missing values
pd.isna(welfare3).sum() # Frequency of missing value
welfare3 = welfare3.dropna(subset = ['ID', 'Year', 'Reg7', 'Gender', 'BirthYear', 'EduLevel', 'MarriType', 'Occupation', 'Income'])
welfare3
```

### 5. Income disparity by occupation

```python
# Find the frequency by occupation
# 2016-2018
welfare1['Occupation'].value_counts()
```

```python
# 2019-2021
welfare2['Occupation'].value_counts()
```

```python
# 2016-2021
welfare3['Occupation'].value_counts()
```

```python
# Matching jobs using job codes
list_job = pd.read_excel('Koweps_Codebook_2021.xlsx', sheet_name = '직종코드')
list_job
```

```python
# Combining a job list to the data
welfare1 = welfare1.merge(list_job, how = 'left', on = 'Occupation')
welfare2 = welfare2.merge(list_job, how = 'left', on = 'Occupation')
welfare3 = welfare3.merge(list_job, how = 'left', on = 'Occupation')
```

```python
# Remove Occupation missing values and output Occupation, list_job
welfare1.dropna(subset = ['Occupation'])[['Occupation', 'list_job']]
welfare2.dropna(subset = ['Occupation'])[['Occupation', 'list_job']]
welfare3.dropna(subset = ['Occupation'])[['Occupation', 'list_job']]
```

```python
# Making an Income avarage table by occupation
# 2016-2021
Job_income3 = \
    welfare3.dropna(subset =['Income'])\
           .groupby(['list_job'], as_index = False) \
           .agg(mean_income = ('Income', 'mean'))
Job_income3
```

```python
# Making a well-earned Job Graph
# Extract Top 10
# 2016-2021
top2021 = Job_income3.sort_values('mean_income', ascending = False).head(10)
top2021
```

```python
# Font settings
import matplotlib.pyplot as plt
plt.rcParams.update({'font.family' : 'Malgun Gothic'})
```

```python
# 2016-2021
sns.barplot(data = top2021, y = 'list_job', x = 'mean_income')
```

```python
# Making a low-income job
# Extract the bottom 10
# 2016-2021
bottom2021 = Job_income3.sort_values('mean_income').head(10)
bottom2021
```

```python
# 2016-2021
sns.barplot(data = bottom2021, y = 'list_job', x = 'mean_income') \
            .set(xlim = (0, 400))
```
